## Data quality checks

### Subtask:
Data Quality Checks and Cleaning


**Reasoning**:
Handle missing values in the 'Postal Code' column by imputing with the mode, check for missing values in other columns and handle appropriately, remove duplicate rows, and verify and correct data types, particularly for categorical columns.



```python
print("Missing values before handling:")
print(df.isnull().sum())

# Handle missing values in 'Postal Code' by imputing with the mode
if 'Postal Code' in df.columns:
    mode_postal_code = df['Postal Code'].mode()[0]
    df['Postal Code'].fillna(mode_postal_code, inplace=True)
    print("\nMissing values in 'Postal Code' handled with mode imputation.")

# Check for missing values in other columns and handle if necessary
# In this dataset, only 'Postal Code' had missing values based on df.info()
# If there were others, we would add logic here based on the column type

print("\nMissing values after handling:")
print(df.isnull().sum())

print("\nNumber of duplicate rows before removal:")
print(df.duplicated().sum())

# Remove duplicate rows
df.drop_duplicates(inplace=True)

print("\nNumber of duplicate rows after removal:")
print(df.duplicated().sum())

print("\nData types before verification:")
df.info()

# Verify and correct data types
# Convert 'Postal Code' to integer type as it represents a code
if 'Postal Code' in df.columns:
    df['Postal Code'] = df['Postal Code'].astype(int)

# Convert potentially categorical columns to 'category' dtype
categorical_cols = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

# Convert date columns to datetime objects
date_cols = ['Order Date', 'Ship Date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], format='%d/%m/%Y')


print("\nData types after verification and correction:")
df.info()

# Check for inconsistencies in categorical data (basic check for unique values)
print("\nUnique values in categorical columns (sample):")
for col in categorical_cols:
    if col in df.columns:
        print(f"\n'{col}':")
        display(df[col].value_counts().head()) # Display top few for brevity
```

## Outlier handling

### Subtask:
Outlier Handling in Numerical Features


**Reasoning**:
Identify numerical features suitable for outlier analysis, visualize outliers using box plots, and handle outliers using the IQR method.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Identify numerical features (excluding 'Row ID')
numerical_features = ['Sales', 'Postal Code']

# Visualize outliers using box plots
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_features):
    plt.subplot(1, len(numerical_features), i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()

# Handle outliers using the IQR method
for col in numerical_features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap values at the bounds instead of removing outliers
    df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

# Visualize results (optional)
plt.figure(figsize=(12, 6))
for i, col in enumerate(numerical_features):
    plt.subplot(1, len(numerical_features), i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col} after outlier handling')
plt.tight_layout()
plt.show()
```

## Feature engineering

### Subtask:
Feature Engineering: Create new features


**Reasoning**:
Create the new features 'Sales_per_Order' and 'Days_to_Ship', then visualize them.



```python
# Calculate total orders per customer
orders_per_customer = df.groupby('Customer ID')['Order ID'].count()

# Create a mapping from customer ID to total orders
customer_order_mapping = orders_per_customer.to_dict()

# Create 'Sales_per_Order' feature
df['Sales_per_Order'] = df['Customer ID'].map(customer_order_mapping)
df['Sales_per_Order'] = df['Sales'] / df['Sales_per_Order']

# Create 'Days_to_Ship' feature
df['Days_to_Ship'] = (df['Ship Date'] - df['Order Date']).dt.days

# Visualize the new features
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.histplot(df['Sales_per_Order'], kde=True)
plt.title('Distribution of Sales per Order')

plt.subplot(1, 2, 2)
sns.histplot(df['Days_to_Ship'], kde=True)
plt.title('Distribution of Days to Ship')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
sns.scatterplot(x='Sales_per_Order',y='Sales',data=df)
plt.title('Sales vs Sales per order')

plt.subplot(1,2,2)
sns.scatterplot(x='Days_to_Ship',y='Sales',data=df)
plt.title('Sales vs Days to ship')
plt.tight_layout()
plt.show()
```
