## Load data

### Subtask:
Load the dataset from "/content/body_fat_new (1).csv" into a pandas DataFrame.


**Reasoning**:
Import the pandas library and load the dataset into a DataFrame.



```python
import pandas as pd

df = pd.read_csv('/content/body_fat_new (1).csv')
df.head()
```

## Initial data inspection

### Subtask:
Display the first few rows, check the data types, and look for missing values.


**Reasoning**:
Display the first few rows, check data types, and look for missing values in the DataFrame.



```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

## Univariate analysis

### Subtask:
Perform univariate analysis on the dataset, including generating descriptive statistics, histograms, and box plots for numerical columns, and value counts and bar plots for categorical columns.


**Reasoning**:
Generate descriptive statistics for all numerical columns, create histograms and box plots for each numerical column, and generate value counts and a bar plot for the categorical column 'Category'.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate descriptive statistics for numerical columns
print("Descriptive Statistics for Numerical Columns:")
display(df.describe())

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# 2. Create histograms for numerical columns
print("\nHistograms for Numerical Columns:")
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# 3. Create box plots for numerical columns
print("\nBox Plots for Numerical Columns:")
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=df, y=col)
    plt.title(f'Box Plot of {col}')
    plt.ylabel(col)
    plt.show()

# 4. Get value counts for the categorical column 'Category'
if 'Category' in categorical_cols:
    print("\nValue Counts for 'Category' column:")
    category_counts = df['Category'].value_counts()
    display(category_counts)

    # 5. Create a bar plot for the 'Category' column
    print("\nBar Plot for 'Category' column:")
    plt.figure(figsize=(8, 5))
    sns.barplot(x=category_counts.index, y=category_counts.values)
    plt.title('Distribution of Body Fat Categories')
    plt.xlabel('Category')
    plt.ylabel('Count')
    plt.show()
else:
    print("\n'Category' column not found in the dataset.")
```

## Bivariate analysis

### Subtask:
Perform bivariate analysis on the dataset, including generating scatter plots and calculating correlation coefficients for numerical vs. numerical columns, generating box plots or violin plots for numerical vs. categorical columns, and generating cross-tabulations and stacked bar plots for categorical vs. categorical columns.


**Reasoning**:
Calculate and visualize the correlation matrix for numerical columns, then generate scatter plots for pairs of numerical columns and box plots for numerical columns against the 'Category' column.



```python
# 1. Calculate and display the correlation matrix for all numerical columns
print("Correlation Matrix for Numerical Columns:")
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numerical_cols].corr()
display(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()

# 2. Iterate through the numerical columns and create scatter plots
print("\nScatter Plots for Numerical Columns:")
num_numerical_cols = len(numerical_cols)
# Plot only the lower triangle to avoid redundant plots
for i in range(num_numerical_cols):
    for j in range(i + 1, num_numerical_cols):
        col1 = numerical_cols[i]
        col2 = numerical_cols[j]
        plt.figure(figsize=(8, 6))
        sns.scatterplot(data=df, x=col1, y=col2)
        plt.title(f'Scatter Plot of {col1} vs {col2}')
        plt.xlabel(col1)
        plt.ylabel(col2)
        plt.show()

# 3. Iterate through the numerical columns and create box plots against 'Category'
print("\nBox Plots of Numerical Columns by Category:")
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='Category', y=col)
    plt.title(f'Box Plot of {col} by Category')
    plt.xlabel('Category')
    plt.ylabel(col)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

**Reasoning**:
Check if there are other categorical columns besides 'Category' and perform cross-tabulation and stacked bar plots if they exist.



```python
# 4. If there are multiple categorical columns (besides 'Category'), create a cross-tabulation
categorical_cols = df.select_dtypes(include=['object']).columns
other_categorical_cols = categorical_cols.drop('Category', errors='ignore')

if len(other_categorical_cols) > 0:
    print("\nCross-tabulations of 'Category' with other categorical columns:")
    for col in other_categorical_cols:
        cross_tab = pd.crosstab(df['Category'], df[col])
        display(cross_tab)

        # 5. Generate stacked bar plots for each cross-tabulation
        print(f"\nStacked Bar Plot of 'Category' vs {col}:")
        cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6))
        plt.title(f'Category Distribution by {col}')
        plt.xlabel('Category')
        plt.ylabel('Count')
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()
else:
    print("\nNo other categorical columns found besides 'Category'.")
```

## Feature engineering

### Subtask:
Prepare the data for modeling, including handling categorical features and splitting the data into training and testing sets.


**Reasoning**:
Prepare the data for modeling by separating features and target, encoding the target variable, and splitting the data into training and testing sets. Finally, display the shapes of the resulting sets.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 1. Separate features (numerical columns) and target variable ('Category')
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
X = df[numerical_cols]
y = df['Category']

# 2. Encode the categorical target variable 'Category' into numerical labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# 4. Display the shapes of the training and testing sets
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Model selection and training

### Subtask:
Choose a suitable model for classification (e.g., Logistic Regression, RandomForest Classifier) and train it on the training data.


**Reasoning**:
Import the necessary model and train it on the training data.



```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate a RandomForestClassifier
model = RandomForestClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)
```

## Model evaluation

### Subtask:
Evaluate the trained model using appropriate metrics (e.g., accuracy, precision, recall, F1-score).


**Reasoning**:
Evaluate the trained model using appropriate metrics.



```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Predict the categories for the test set
y_pred = model.predict(X_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

# Print the evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision (weighted): {precision:.4f}")
print(f"Recall (weighted): {recall:.4f}")
print(f"F1-score (weighted): {f1:.4f}")
```

## Prediction

### Subtask:
Use the trained model to predict the category based on numerical input data.


**Reasoning**:
Use the trained model to predict the category based on numerical input data, inverse transform the predictions to original categorical labels, and display the first few predicted categorical labels.



```python
# Use the trained model's .predict() method to make predictions on the test features, X_test.
y_pred = model.predict(X_test)

# Inverse transform y_pred back to the original categorical labels
y_pred_categorical = label_encoder.inverse_transform(y_pred)

# Display the first few predicted categorical labels
print("First few predicted categorical labels:")
display(y_pred_categorical[:10])
```
