## **3. Data Quality Checks**

#### **Checking Missing Values**

```python
#missing value haru check gareko
df.isnull().sum()
```

From this, we get to know that there's 11 missing values in column - "Postal Code".




#### **Dropping Irrelevant Columns**

```python
#removing "Postal Code" column as it is similar to column:  city and is
#irrelevant here

df.drop(columns=["Postal Code"],inplace=True)

#removing "Country" column as all these orders took place only in "United States"
df.drop(columns=["Country"],inplace=True)
```

```python
# #removing "Order ID","Customer ID" and "Product ID" columns as they remain irrelevant here.
df.drop(columns=["Order ID","Customer ID","Product ID"],inplace=True)
```

#### **Checking for Duplicates**

```python
#Checking for duplicates
df.duplicated().sum()
```

No duplicates were found.

```python
# Checking missing values and remaining rows after quality check
df.isnull().sum()
```

####**Verify Data Types**

```python
#Changing 'Ship Date' and 'Order Date' to datetime format and checking data type
df["Ship Date"] = pd.to_datetime(df["Ship Date"], format='%d/%m/%Y')
df["Order Date"] = pd.to_datetime(df["Order Date"], format='%d/%m/%Y')
df.info()
```

```python
#Changing columns with limited categorical value to category type
df["Category"] = df["Category"].astype('category')
df["Sub-Category"] = df["Sub-Category"].astype('category')
df["Ship Mode"] = df["Ship Mode"].astype('category')
df["State"] = df["State"].astype('category')
df["Region"] = df["Region"].astype('category')
df["Segment"] = df["Segment"].astype('category')
df.info()
```

##**4. Outliers Detection and Handling**

```python
#plotting box plot for "Sales" column
plt.figure(figsize=(10,5))
sns.boxplot(df["Sales"])
plt.title("Box Plot of Sales")
```

```python
#outlier handling: remove extreme "Sales" values
q1 = df['Sales'].quantile(0.25)
q3 = df['Sales'].quantile(0.75)
iqr = q3 - q1
upper_limit= q3 + 1.5 * iqr
lower_limit= q1 - 1.5 * iqr
df = df[(df['Sales'] >= lower_limit) & (df['Sales'] <= upper_limit)]
print(upper_limit)
print(lower_limit)
```

## **5.Feature Engineering**

```python
# Create 'Shipping Time' feature
df["Shipping Time"] = (df["Ship Date"] - df["Order Date"]).dt.days
```
