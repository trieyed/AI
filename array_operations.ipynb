```python
import numpy as np
```

```python
#creating a scaler
scaler = 7
print(f"Scaler:{scaler}")

#creating a vector
vector = np.array([1,2,3])
print(f"Vector:{vector}")

#Scaling vector through a scalar aka-creating scaled vector
scaled_vector = scaler*vector
print(f"Scaled Vector:{scaled_vector}")
```

```python
vector = np.array([4,5,6])
scaled_vector = 7*vector
print(f"Scaled Vector:{scaled_vector}")
```

```python
vector1 = np.array([10,20])
vector2 = np.array([5,15])
vector3 = vector1 + vector2
print(f"Sum of vector1 and vector 2:{vector3}")

vector4= vector1 - vector2
print(f"Difference of vector1 and vector 2:{vector4}")
```

```python
# Matrix Creation and Basic Operations
matrix1 = np.array([2,4],[5,7])
matrix2 = np.array([6,8],[9,11])

print(matrix1)
print(matrix2)

print(matrix1+matrix2)
print(matrix1-matrix2)
```
