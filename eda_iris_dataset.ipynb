```python
#Visualize categorical count
plt.figure(figsize=(10,5))
sns.countplot(x="species",data=df)
plt.show()
```



1.   The given bar chart describes the number of various species: iris setos,   iris versicolor, and iris verginica.

2.   The result shows number of each species in the data is 50 each.

```python
#Correlation Matrix(for numerical columns)
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr,annot = True ,cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```



1.   The given image shows a heatmap of the numerical features of iris dataset
2.   Blue represents negative correlation and its intensity represents how negative it is.
3. Red represents positive correation and its intensity represents how positive of correlation it is.
4. For example: sepal length and sepal petal length has correlation -0.43 meaning they are negatively related.
