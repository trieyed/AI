```python
# Task 4: Calculate the probability of drawing a heart or a club from a deck (13
# hearts, 13 clubs, mutually exclusive). Print the result.

favorable_hearts = 13
favorable_clubs = 13
total_cards = 52

# Probability of drawing a heart
prob_heart = favorable_hearts / total_cards

# Probability of drawing a club
prob_club = favorable_clubs / total_cards

# Since drawing a heart and drawing a club are mutually exclusive events,
# the probability of drawing a heart OR a club is the sum of their probabilities.
prob_heart_or_club = prob_heart + prob_club
prob_heart_or_club
```

```python
# Task 5: Calculate the probability of rolling a 6 twice in a row with a six-sided
# die. Print the result.
# # Question: What is the probability of getting head on a coin toss and 7 on a 6-sided die roll?

# Probability of getting a 6 on a single roll
prob_six = 1/6

# Probability of getting a 6 twice in a row (independent events)
prob_two_sixes = prob_six * prob_six
print(f"The probability of rolling a 6 twice in a row is: {prob_two_sixes}")

# Probability of getting a head on a coin toss
prob_head = 1/2

# Probability of getting a 7 on a 6-sided die roll
prob_seven = 0/6 # It's impossible to get a 7

# Probability of getting a head and a 7 (independent events)
# Note: If the question meant getting a 1 on the die, use 1/6
prob_head_and_seven = prob_head * prob_seven
print(f"The probability of getting a head on a coin toss and a 7 on a 6-sided die roll is: {prob_head_and_seven}")
```

```python
# Task 6: Calculate the probability of drawing two red cards without replace-
# # ment from a deck (26 red cards, 52 total). Print the result.

# Total number of red cards
red_cards = 26
# Total number of cards in the deck
total_cards = 52

# Probability of drawing the first red card
prob_first_red = red_cards / total_cards

# After drawing one red card without replacement, the number of red cards decreases by 1
remaining_red_cards = red_cards - 1
# The total number of cards in the deck also decreases by 1
remaining_total_cards = total_cards - 1

# Probability of drawing the second red card given the first was red
prob_second_red_given_first_red = remaining_red_cards / remaining_total_cards

# Probability of drawing two red cards without replacement is the product of the two probabilities
prob_two_red_no_replacement = prob_first_red * prob_second_red_given_first_red

print(f"The probability of drawing two red cards without replacement is: {prob_two_red_no_replacement}")
```
